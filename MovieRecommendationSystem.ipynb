```python
# @title
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('movies.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
features=['genres','keywords','original_language','title','cast','director']
for feature in features:
  df[feature]=df[feature].fillna('')
```

```python
def combined_features(row):
  return row['title']+','+row['genres']+','+row['keywords']+','+row['original_language']+','+row['title']+','+row['cast']+','+row['director']
df['combined_features']=df.apply(combined_features,axis=1)
df['combined_features']
```

```python
tfid=TfidfVectorizer()
tfidv=tfid.fit_transform(df['combined_features'])
tfidv.toarray()
```

```python
tfidv.shape
```

```python
cosine_sin=cosine_similarity(tfidv)
cosine_sin
```

```python
cosine_sin.shape
```

```python
movie=input('Enter Movie Name: ')
def get_index(mn):
  return df[df.title==mn].index[0]
mi=get_index(movie)
mi
```

```python
sm=list(enumerate(cosine_sin[mi]))
print(sm)
```

```python
sorted_sm=sorted(sm,key=lambda x:x[1],reverse=True)
print(sorted_sm)
```

```python
def get_info(index):
  return df[df.index==index]['title'].values[0]+': '+df[df.index==index]['cast'].values[0]
i=0
for movie in sorted_sm:
  print(get_info(movie[0]))
  i=i+1
  if i>10:
    break
```
